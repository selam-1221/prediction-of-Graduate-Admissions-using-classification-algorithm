```python
df=df.rename(columns={'Serial No.':'Serial_No','GRE Score':'GRE_Score','TOEFL Score':'TOEFL_Score','University Rating':'University_rating','LOR ':'LOR'
                     ,'chance of admit ':'Chance_of_Admit'})
```

```python
# Replace 'ChanceofAdmit' with the correct column name 'Chance of Admit'
target_variable = 'Chance_of_Admit'

# Analyze the distribution of the target variable
target_distribution = df[target_variable].value_counts(normalize=True)
print(f"\nDistribution of the Target Variable ({target_variable}):")
print(target_distribution)

# Check if the target variable is balanced or imbalanced
is_imbalanced = target_distribution.max() > 0.8 or target_distribution.min() < 0.2
print(f"\nIs the Target Variable Imbalanced? {is_imbalanced}")
```

```python
df['University_rating'].value_counts().plot(kind='pie',autopct='%1.0f%%',figsize=(8,6),
                                            title="University Rating Count Plot")
plt.show()
```

```python
columns_to_plot = ['University_rating','TOEFL_Score']

# Create subplots
fig, axs = plt.subplots(1, len(columns_to_plot), figsize=(8 * len(columns_to_plot), 6))

# Plot each column as a pie chart
for i, column in enumerate(columns_to_plot):
    df[column].value_counts().plot(kind='pie', autopct='%1.0f%%', ax=axs[i], title=f"{column} Count Plot")

# Show the plots
plt.show()
```
